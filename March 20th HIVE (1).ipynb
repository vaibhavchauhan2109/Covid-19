Spark SQL Hive Integration

```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.config("hive-query","value").enableHiveSupport().getOrCreate()
```

```
%sql
CREATE TABLE IF NOT EXISTS HiveQ(
          Key int,
          Value string)
          ROW FORMAT DELIMITED
          FIELDS TERMINATED BY ','
```

```
spark.sql("LOAD DATA INPATH '/FileStore/tables/val.txt' INTO TABLE HiveQ")
```

```
%sql 
SELECT * FROM HiveQ
```

```
spark.sql('SELECT COUNT(*) from HiveQ').show()
```

```
sqlDF = spark.sql('Select Key, Value from HiveQ Where Key > 5')
```

```
stringDS = sqlDF.rdd.map(lambda row : 'Key: %d, Value: %s' %(row.Key,row.Value) )
```

```
stringDS
```

```
for i in stringDS.collect():
  print(i)
```

```
from pyspark.sql import Row
Record = Row("Key","Value")
recordsDF = spark.createDataFrame([Record(i, "val_" + str(i))for i in range(1,10)])
recordsDF.createOrReplaceTempView("records")
```

```
%sql
Select * from records
```

```
recordsDF.show()
```

```
sql("SELECT * FROM records r JOIN HiveQ s ON r.Key = s.Key").show()
```

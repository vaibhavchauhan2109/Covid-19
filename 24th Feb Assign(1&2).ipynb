###An array contains n numbers ranging from 0 to n-1. There are some numbers duplicated in the array.
How do you find a duplicated number in the array?

```
lis=[int(x) for x in input().split()]
newlist=sorted(lis)
dic=dict(enumerate(newlist))
print(dic)
dup=[]
for i in range(len(newlist)-1):
    if newlist[i] == newlist[i+1]:
        dup.append(newlist[i])
print(dup)
```

###Assignment 4: For a string. There can have multiple special characters inside the string. Write a program to reverse the given string. The position of the special characters should not be changed.

```
str1='Hi my name is vaib @#$hg how are  you* doing'
special='!@#$%^&*'
lis=[]
for i in str1[::-1]:
    for j in i:
        if j not in special:
            lis.append(j)
        else:
            continue
for i in str1:
    for j in i:
        if j in special:
            lis.insert(str1.index(j),j)
#print(lis)
result = ''.join(lis)
  
print(result) 
            
            
```
